
2. Найти нормы векторов (4,2,4) и (12,3,4) и угол между ними.



```python
x = np.array([4, 2, 4])
```

Манхэттенская норма вектора x

```python
np.linalg.norm(x, ord=1)
```

Евклидова норма вектора x

```python
norm_x = np.linalg.norm(x, ord=2)
norm_x
```

```python
y = np.array([12, 3, 4])
```

Манхэттенская норма вектора y

```python
np.linalg.norm(y, ord=1)
```

Евклидова норма вектора y

```python
norm_y = np.linalg.norm(y, ord=2)
norm_y
```

Угол между векторами х и у

```python
cos_phi = np.dot(x, y) / norm_x / norm_y
cos_phi # косинус угла
```

```python
np.degrees(np.arccos(cos_phi)) # угол в градусах
```

4. Какие из нижеперечисленных векторов образуют ортонормированный базис в линейном пространстве R3:



а)Ортонормированы, т.к они ортогональны и нормированы.

```python
a = np.array([1,0,0])
b = np.array([0,0,1])
```

```python
np.dot(a,b) # ортогональны
```

```python
np.dot(a,a) # нормированы
```

```python
np.dot(b,b) # нормированы
```

```python
def is_ortonorm(a,b,c):
    if np.dot(a,b) == 0 and np.dot(c,b) == 0 and np.dot(c,a) == 0:
        if np.dot(a,a) == 1 and np.dot(b,b) == 1 and np.dot(c,c) == 1:
            print('Вектора a,b,c ортонормированы')
        else:
            print('Какие-то из векторов не нормированы: ', np.dot(a,a), np.dot(b,b), np.dot(c,c))
            
    else:
        print('Вектора a,b,c не ортогональны')
```

б) Не ортонормированы

```python
a = np.array([1/np.sqrt(2), -1/np.sqrt(2), 0])
b = np.array([1/np.sqrt(2), 1/np.sqrt(2), 0])
c = np.array([0,0,1])
```

```python
is_ortonorm(a,b,c)
```

в) Не ортонормированы

```python
a = np.array([1/2, -1/2, 0])
b = np.array([0, 1/2, 1/2])
c = np.array([0,0,1])
```

```python
is_ortonorm(a,b,c)
```

г) Ортонормированы

```python
a = np.array([1, 0, 0])
b = np.array([0, 1, 0])
c = np.array([0,0,1])
```

```python
is_ortonorm(a,b,c)
```
